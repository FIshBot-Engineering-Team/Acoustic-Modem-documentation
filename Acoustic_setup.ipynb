# **Acoustic Modem Setup Guide**

---------------------------

### Preliminarty Development:

##### Overview: 
This document outlines the steps taken toward developing the acoustic modem for KnifeBot. The modem will support long-range underwater acoustic sensing and communication, with hydrophones used to improve signal clarity and help ensure consistent sound quality.

1. Required Devices
      - RaspberryPi5 or equivalent microcontrollers( esp32, STM32, Etc. ) 
      - Keyboard
      - Mouse
      - Monitor
2. Software Required
      - Minimum: Windows 7 (Windows 11 in our case) 
      - Linux Distribution (Ubunutu 22.04 or newer)
      - PuTTy (terminal emulator and network client software)
        - This will be used in both Windows and Linux operating systems

------------------------------

## **Testing via Windows**

This section briefly covers the general steps for setting up the Popoto modem on a Windows system.

### Step 1: Hardware connections
1. Connect and power the Popoto modem battery.
2. Connect the data cable from the Popoto acoustic modem to the digital board.
3. Connect a USB Type-C cable from the digital board to the user's laptop/desktop.

- We used a Type-C to Type-A cable, but Type-C to Type-C also works.

### Step 2: Install PuTTY
Download PuTTY from the Microsoft Store:  
https://apps.microsoft.com/detail/XPFNZKSKLBP7RJ?hl=en-US&gl=US&ocid=pdpshare

### Step 3: Open Device Manager
Click the Windows button and search **Device Manager**.



### Step 3.1: Find the COM port
Expand **Ports (COM & LPT)** and note the COM port number (example: COM3).




### Step 4: Configure PuTTY (Serial)

Open PuTTY and use the settings shown in the image below. Once the application loads, set **Connection type** to **Serial**, then enter your devices COM port (from the previous step) in the **Serial line** field. Set **Speed** to **115200**. Optionally, use **Saved Sessions** to save this configuration for faster setup next time. Click **Open**a separate terminal window will appear.
